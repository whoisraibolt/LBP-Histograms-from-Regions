```python
def LBP_histograms_regions(path, use_circular_neighborhood=True):
    # Number of points
    P = 8
    
    # Radius
    R = 2
    
    # It receives a vector with 255 positions (0 to 255)
    bins = np.array(range(256))
    # print'\nBins:',bins

    # Opens a image
    img = Image.open(path)

    # Shows the width and height of the image and the color indexing mode
    print img.size, img.mode
    
    # Transforms the image into an array
    image = np.asarray(img)
    # print'\nImage array:',image
    
    # Regular mesh is 7x7 which gives us 49 regions, each region contains 18x18 pixels (window size)
    
    # Define Grid (k * K)
    k = 7
    
    # Define the Window size
    window_size_row = image.shape[0] / k
    window_size_column = image.shape[1] / k
    # print 'Window size row:',row
    # print 'Window size column:',column

    store_all = []
    count_region = 0

    # Scroll from the upper left corner of the image
    # 'Cutting' the window and calculating the LBP histogram
    # i = Intensity
    # r = Region
    for i in range(0, image.shape[0] - window_size_row, window_size_row):
        # Shows in which row of the region we are navigating (1 to 7)
        # Remember that each row occupies 7 rows
        print('\nWindow row: '+str(((i / window_size_row) + 1)))
        
        # With each row, runs 18 columns
        for r in range(0, image.shape[1] - window_size_column, window_size_column):
            # Shows in which column of the region we are navigating (1 to 7)
            # # Remember that each column occupies 7 columns
            print('\nWindow column: '+str(((r / window_size_column) + 1)))

            # Counters of regions
            count_region = count_region + 1
            
            # Region
            region = image[i:i + window_size_row, r:r + window_size_column]
            
            # Shows the 'cropped' region
            print('\nRegion: '+str(count_region))
            print region
            
            
            if use_circular_neighborhood:
                # Compute the LBP circular_neighborhood representation of the regions of each image,
                # and then use this to build the histogram
                
                # The uniform  method indicates that we are computing the rotation
                # and grayscale invariant form of LBPs.
                LBP = feature.local_binary_pattern(region, P, R, method="uniform")
                
                # Histogram
                (hist, _) = np.histogram(LBP, bins=np.arange(0, P + 3), range=(0, P + 2))
            else:
                # Histogram
                (hist, _) = np.histogram(region, bins=bins)
    
            # Shows the Histogram
            print '\nHistogram: '
            print hist
            
            # Join all n regions histograms of each image
            store_all.append(hist)
            
    # Shows
    print('\nAll n region histograms: '+str(store_all))
    
    return store_all
```

```python
histogram_dictionary = {}

# Call the functions
for fn in glob.glob('*.jpg'):
    print '\nImage:',fn

    # Gets the file name without the extension
    f = glob.os.path.splitext(fn)[0]
    
    # If your images are already the result of the basic LBP operator (in this case), set the [variable] as False,
    # but if you want to take your images the LBP circular (P, R) neighborhood,set the [variable] as True
    histogram_dictionary["{0}".format(f)] = LBP_histograms_regions(path=fn,
                                                                   use_circular_neighborhood=False)
```

```python
def chi_square(histogram1, histogram2, weight):
    #Computes the chi-square distance between two histograms (or histogram sequences)

    x2w = 0
    store_all = []
    
    for i in range(0, histogram1.shape[0]):
        if histogram1[i] != histogram2[i]:
            x2w += weight * int(((histogram1[i] - histogram2[i])**2) / (histogram1[i] + histogram2[i]))
   
    # 49 x2w of each image
    store_all.append(x2w)
    
    return store_all
```
